```python
n = 1024
a = cl.clrandom.rand(queue, n, dtype=np.float32)
b = cl.clrandom.rand(queue, n, dtype=np.float32)
```

## Instruction tagging

```python
%%transformed_fortran_kernel
subroutine rot_norm(out, alpha, out2, inp, inp2, n)
  implicit none
  real*8 a, b, r, out(n), out2(n), inp(n), inp2(n)
  real*8 alpha
  integer n, i

  do i = 1, n
    !$loopy begin tagged: input
    a = cos(alpha)*inp(i) + sin(alpha)*inp2(i)
    b = -sin(alpha)*inp(i) + cos(alpha)*inp2(i)
    !$loopy end tagged: input

    r = sqrt(a**2 + b**2)
    a = a/r
    b = b/r

    out(i) = a
    out2(i) = b
  end do
end

!$loopy begin
!RESULT = lp.parse_fortran(SOURCE, FILENAME)
!$loopy end
```

```python
for insn in lp.find_instructions(rot_norm, "*$input"):
    print insn.assignees_and_indices()[0]
```

## Substitution rules: Semantics

```python
knl = lp.make_kernel(                                                        
    "{[i,j,n,n2]: 0<=i,j<npart and 0<=n,n2<3}",                          
    """                                                                  
    grav_force(m, M, r) := -66.742*m*M/r**2                              

    <> radc = sqrt(sum(n, (x[i,n]-center[n])**2))                        
    <> rad_j = sqrt(sum(n2, (x[i,n2]-x[j,n2])**2))                       

    force[i] = grav_force(mass[i], massc, radc) + \
        sum(j, grav_force(mass[i], mass[j], rad_j))
    """)
```

```python
knl_lines = str(knl).split("\n")
print("\n".join(knl_lines[23:]))
```

Substitution rules are just named subexpressions, optionally with parameters.

## Substitution Rules: Purpose I

Substitution rules help operate on subexpressions, such as for precomputation:

```python
%%transformed_fortran_kernel
subroutine diff(out, u, n)
  implicit none
  real*4 out(n), u(n+1)
  integer n, i

  do i = 1, n
    out(i) = u(i+1) - u(i)
  end do
end

!$loopy begin
!RESULT = lp.parse_fortran(SOURCE, FILENAME)
!$loopy end
```

```python
diff = lp.set_options(diff, write_cl=True)
diff(queue, u=a)
```

```python
diff_split = diff
diff_split = lp.split_iname(diff_split, "i", 16)
diff_split = lp.assume(diff_split, "n mod 16 = 0")

#diff_split = lp.extract_subst(diff_split, "u_acc", "u[j]", parameters="j")
#diff_split = lp.precompute(diff_split, "u_acc", "i_inner", default_tag=None)
#print diff_split

diff_split(queue, u=a)
```

## Substitution Rules: Purpose II

They also help with selection of subexpressions:

```python
knl = lp.make_kernel(
        "{[i]: 0<=i<n}",
        """
            f(x) := x*a[x]
            g(x) := 12 + f(x)
            h(x) := 1 + g(x) + 20*g$two(x)

            a[i] = h$one(i) * h$two(i)
            """)
```

```python
knl = lp.expand_subst(knl, "g$two < h$two")
print knl
```

## Substitution Rules: Creation by Unification

How are substitution rules created, given that Fortran codes do not contain them?

```python
%%fortran_kernel
subroutine unif(a, b, n)
  implicit none
  real*4 a(n), b(n)
  integer n, i

  do i = 1, n
    a(i) = 23*b(i)**2 + 25*b(i)**2
  end do
end
```

```python
print(unif)
```

```python
unif = lp.extract_subst(unif,
  "bsquare", "alpha*b[i]**2", parameters=("alpha",))
print(unif)
```

## Substitution Rules: Creation by Extraction of Temporary

```python
%%fortran_kernel

subroutine temp_subst(out, inp, n)
  implicit none
  integer i, n
  real*4 inp(n), out(n), a(n)

  do i = 1, n
    a(i) = 6*inp(i)
    out(i) = 5*a(i)
  end do
end
```

```python
print temp_subst
```

```python
temp_subst = lp.temporary_to_subst(temp_subst, "a")
print(temp_subst)
```
