## Kernel-level processing

```python
%%transformed_fortran_kernel
subroutine add_term_1(out, a, n)
  implicit none

  real*4 a, out(n)
  integer n, i

  do i = 1, n
    out(i) = out(i) + a
  end do
end

subroutine add_term_2(out, b, n)
  implicit none

  real*4 out(n),b(n)
  integer n, i

  do i = 1, n
    out(i) = out(i) + 2*b(i)
  end do
end

!$loopy begin
!
! fill, twice = lp.parse_fortran(SOURCE)
! knl = lp.fuse_kernels((fill, twice))
! RESULT = [knl]
!
!$loopy end
```

## More involved control flow

```python
%%transformed_fortran_kernel

subroutine fill(out, inp, n)                               
  implicit none                                                  

  real*4 alpha, beta, out(n), inp(n)                           
  integer n, i, j                                                

  do i = 1, n                                                    
    alpha = inp(i)                                                   
    if (alpha.ge.3) then                                             
        beta = 2*alpha                                                
        do j = 1,3                                               
            beta = 3 * beta
        end do                                                   
        out(i) = 5*beta
    else                                                         
        out(i) = 4*alpha
    endif                                                        
  end do                                                         
end

!$loopy begin
!RESULT = lp.parse_fortran(SOURCE, FILENAME)
!$loopy end
```

```python
fill = lp.set_options(fill, write_cl=True)
fill(queue, inp=a)
```
