```python
def quicksort(l):
    if len(l) <= 1:
        # already sorted
        return l
    
    pivot = l[0]
    
    left = [item for item in l[1:] if item <= pivot]
    right = [item for item in l if item > pivot]

    return quicksort(left) + [pivot] + quicksort(right)
```

```python
from random import randrange
mylist = [randrange(0, 100) for i in range(20)]
```

```python
quicksort(mylist)
```
