# What is ibvp?

Ambition: A data model for 

* PDEs
* Boundary Value Problems
* Initial Boundary Value Problems

----
Idea:

* Describe the problem you'd like solved
* Push a button, export to umpteen PDE solvers

------------

Currently: Prototyped using Proteus

```python
import ibvp.sym as sym
from ibvp.language import PDESystem
from ibvp.target.proteus import generate_proteus_problem_file

u = sym.Field("u")

eqns = sym.join(sym.d_dt(u) + sym.d_dx(u))
print(sym.pretty(eqns))
```

```python
print(generate_proteus_problem_file(
    PDESystem(
        ambient_dim=1,
        pde_system=eqns,
        unknowns=[u],
        ),
    "Advection"))
```

------------

Components:

* PDEs/I?BVPS: [Top level](https://github.com/ibvp/ibvp/blob/master/ibvp/language/__init__.py)
*  Symbolic expressions based on an extension of pymbolic: [Expression language](https://github.com/ibvp/ibvp/blob/master/ibvp/language/symbolic/primitives.py)
* Proteus target (independent of data model): [Translator](https://github.com/ibvp/ibvp/blob/master/ibvp/target/proteus/__init__.py)
    * Based on transformation framework expressed in pymbolic: [Mappers](https://github.com/ibvp/ibvp/blob/master/ibvp/target/proteus/__init__.py)
