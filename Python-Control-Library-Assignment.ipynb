1. Create a System Model with Transfer function $$\frac{1}{s^2 + 1.92 s + 5.76}$$

```python
import numpy as np
from control.matlab import *
sys = tf([1.], [1., 1.92, 5.76])
sys
```

2. Convert the System Transfer Function model into Statespace Model

```python
sys2=tf2ss(sys)
sys2
```

3. Plot the Step Response of the above Model. Save the step response as a figure.

```python
# Do the Step response
T = np.arange(0,10,0.001)
yout, T = step(sys, T)

# Plot the step response
import matplotlib
import matplotlib.pyplot as plt

%config InlineBackend.figure_format = 'retina'

fig, ax = plt.subplots()
ax.plot(T,yout)

ax.set(xlabel='time', ylabel='amplitude',
       title='Step Response of a 2nd Order System')
# ax.grid()

fig.savefig("Step Response.png")
plt.show()
```

4. Plot the Impulse Response of the above model. Give a Title to the plot, use suitable X-axis labels and Y-axis labels.

```python
# Impulse Response
yout, T = impulse(sys,T)

# Plot the response
import matplotlib
import matplotlib.pyplot as plt

%config InlineBackend.figure_format = 'retina'

fig, ax = plt.subplots()
ax.plot(T,yout)

ax.set(xlabel='time', ylabel='amplitude',
       title='Impulse Response of a 2nd Order System')
# ax.grid()

fig.savefig("Impulse Response.png")
plt.show()
```
