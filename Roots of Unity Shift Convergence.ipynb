```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 15

eigenvalues = np.exp(2*np.pi*1j*np.arange(n)/n)

sigma = 1.2 - 0.3j
```

```python
plt.plot(eigenvalues.real, eigenvalues.imag, "o")
plt.plot(sigma.real, sigma.imag, "ro")

plt.gca().set_aspect("equal")
```

```python
shift_inverted = 1/(eigenvalues-sigma)

plt.plot(shift_inverted.real, shift_inverted.imag, "o")
```
