```python
#-------------------------------------------------------------------------------
# Name:        plot_gate_signals
# Purpose:     plots switch signals for power electronics converters
#
# Author:      Tom
#
# Dependencies: Anaconda
#
# Created:     06/27/2014
# Updates:     06/28/2014: shading
# Copyright:   (c) Tom 2014
# Licence:     open license
#-------------------------------------------------------------------------------
 
%pylab inline
from matplotlib.path import Path
import matplotlib.patches as patches
import matplotlib.pyplot as plt
 
# initial state, toggle, toggle, ...
 
#----------------------------------------------------------------
# User defined parameters
#
line_color = 'k'
 

title_text = 'Gate signals for phase-shifted 3-level dc-dc converter'
 
s1 = ['off',50,100,150,200]
s4 = ['on',50,100,150,200]
s2 = ['off',50,60,150,160]
s3 = ['on',10,100,110,200,210]
 
s1s = [50,60,150,160]
s2s = []
s3s = []
s4s = [0,10,100,110,200,210]
 
xlabel_text = ['Ts/2', 'Ts','3/2 Ts', '2 Ts','']
xlabel_loc = [50,100,150,200,250]
ylabel_text = ['S1','S2','S3','S4']
 
switch = [s1,s2,s3,s4]
shading = [s1s,s2s,s3s,s4s]

 
title_text = 'Gate signals for synchronous buck dc-dc converter'
 
dt = 2
s1 = ['off',50+dt,100,150+dt,200]
s2 = ['on',50,100+dt,150,200+dt]
 
s1s = []
s2s = []
 
xlabel_text = ['Ts/2', 'Ts','3/2 Ts', '2 Ts','']
xlabel_loc = [50,100,150,200,25]
# Figure height = No. of signals + 1
plt.figure(num=None, figsize=(9, shape(switch)[0]+1), dpi=600, facecolor='w', edgecolor='w')
              
# plt.figure(num=None, figsize=(9, shape(switch)[0]+1), dpi=600, facecolor='w')
ax = plt.axes()
 
# Gate signal
for switch_number in range(shape(switch)[0]):
    s1 = switch[switch_number];
    if s1[0] == 'on':
        level = 1
    else:
        level = 0
 
    for k in range(len(s1)-1):
        if k == 0:
            plot([0,s1[k+1]],[level-switch_number*2,level-switch_number*2],linewidth=4, color=line_color)  # first horizontal lines
        plot([s1[k+1], s1[k+1]],[-switch_number*2,1-switch_number*2],linewidth=4, color=line_color)  # vertical lines
        level = 1 - level
        if k < (len(s1)-2):
            plot([s1[k+1],s1[k+2]],[level-switch_number*2,level-switch_number*2],linewidth=4, color=line_color)  # horizontal lines that connect switch transitions
        else:
            plot([s1[k+1],230],[level-switch_number*2,level-switch_number*2],linewidth=4, color=line_color)  # last horizontal line
 
# Shading
for switch_number in range(shape(shading)[0]):
    s1 = shading[switch_number];
    for shade in range(len(s1)/2):
        verts = [
        (s1[shade*2], 0-2*switch_number), # left, bottom
        (s1[shade*2], 1-2*switch_number), # left, top
        (s1[shade*2+1], 1-2*switch_number), # right, top
        (s1[shade*2+1], 0-2*switch_number), # right, bottom
        (0., 0.), # ignored
        ]
        ax.add_patch(patches.PathPatch(Path(verts,[Path.MOVETO,Path.LINETO,Path.LINETO,Path.LINETO,Path.CLOSEPOLY]), facecolor=shade_color, lw=0))
 
ylim([1-2*shape(switch)[0],2])  # y-axis limit
ax.xaxis.grid() # show vertical lines only
 
# Remove the tick along the left and right edges
plt.tick_params(\
    axis='y',          # changes apply to the y-axis
    which='both',      # both major and minor ticks are affected
    left='off',      # ticks along the bottom edge are off
    right='off',         # ticks along the top edge are off
    labelbottom='off') # labels along the bottom edge are off
 
# White frame
sides = ['bottom','left','top','right']
for side in sides:
    ax.spines[side].set_color('white')
 
# Set ticks and title
ylabel_loc = linspace(0.5,2.5-2*shape(switch)[0],shape(shading)[0])
xticks(xlabel_loc, xlabel_text,fontsize=16)
yticks(ylabel_loc, ylabel_text,fontsize=16)
plt.suptitle(title_text,fontsize=14)
 
# Save the figure
savefig(title_text + '.pdf', bbox_inches='tight')
savefig(title_text + '.png', bbox_inches='tight')
 
# no more
```
