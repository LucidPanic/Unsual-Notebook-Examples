```python
from __future__ import division, print_function

import numpy as np

import ibvp.sym as sym
from ibvp.language import PDESystem
from ibvp.target.proteus import generate_proteus_problem_file
```

```python
ambient_dim = 2    
u = sym.Field("u")
```

### Heat equation

```python
eqns = sym.join(
        sym.d_dt(u)
        - sym.div(sym.grad(u))
        )

print(sym.pretty(eqns))
```

```python
print(generate_proteus_problem_file(
        PDESystem(
            ambient_dim=ambient_dim,
            pde_system=eqns,
            unknowns=[u],
            ),
        "Heat"))
```

### Burgers

```python
vec = np.array([1.0, 2.0])

eqns = sym.join(
        sym.d_dt(u)
        + sym.div(vec * u**2)
        - sym.div(sym.grad(u))
        )

print(sym.pretty(eqns))
```

```python
print(generate_proteus_problem_file(
        PDESystem(
            ambient_dim=ambient_dim,
            pde_system=eqns,
            unknowns=[u],
            ),
        "Burgers"))
```
