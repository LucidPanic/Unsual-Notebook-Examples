## Gradient descent

Consider the least-squares problem of minimizing
$$ \|A\vec y - \vec b\|_2^2. $$

```matlab
A = randn(8,2)+1.4
b = 20*ones(8,1)
```

We are trying to minimize the function 
$$ f(y_1,y_2) = \left\| A \begin{bmatrix} y_1 \\ y_2 \end{bmatrix} - \vec b \right\|_2^2.$$

```matlab
f = @(y1,y2)  norm(A*[y1;y2]-b)^2
```

```matlab
f(1,1)
```

```matlab
Y1 = linspace(-10,10,100);
Y2 = linspace(-8,16,100);
[Y1,Y2] = meshgrid(Y1,Y2);
Z = Y1;
for i = 1:size(Y1,1)
    for j = 1:size(Y1,2)
        Z(i,j) = f(Y1(i,j),Y2(i,j));
    end
end
low = norm(A*(A\b)-b)^2;
curves = exp(linspace(log(low)-.1,log(low)+5,60));
```

```matlab
contour(Y1,Y2,Z,curves)
title('The level curves of f')
xlabel('y1')
ylabel('y2')
```

```matlab
y = [-10;8];
```

```matlab
y1s = [y(1)];
y2s = [y(2)];
contour(Y1,Y2,Z,curves)
title('The level curves of f')
xlabel('y1')
ylabel('y2')
hold on
plot(y1s,y2s,'ko','MarkerFaceColor',[0,0,0])
```

```matlab
c = A'*b;
r = A'*(A*y) - c;
v = A*r;
h = (r'*r)/(v'*v);
y = y - h*r;
y1s = [y1s,y(1)];
y2s = [y2s,y(2)];
contour(Y1,Y2,Z,curves)
title('The level curves of f')
xlabel('y1')
ylabel('y2')
hold on
plot(y1s,y2s,'ko','MarkerFaceColor',[0,0,0])
```

## In higher dimensions

```matlab
A = randn(18,6)+1.4;
b = 20*ones(18,1);
c = A'*b;
```

```matlab
y = zeros(6,1);
```

```matlab
for i = 1:1000
    r = A'*(A*y) - c;
    v = A*r;
    h = (r'*r)/(v'*v);
    y = y - h*r
    if norm(h*r) < 1e-10
        break
    end
end
```
