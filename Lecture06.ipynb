## Rounding errors

Solve $0.00001x^2 + 20x + 1 = 0$

$$ x = \frac{-20 \pm \sqrt{400 - 4(0.0001)}}{0.00002}$$

```matlab
format long
a = 0.00001; b = 20.; c = 1;
x = (-b - sqrt(b^2 - 4*a*c))/(2*a)
a*x^2 + b*x + c
```

```matlab
format long
a = 0.00001; b = 20.; c = 1.;
x = (-b + sqrt(b^2 - 4*a*c))/(2*a)
a*x^2 + b*x + c
```

```matlab
a = 0.00001; b = 20.; c = 1.;
x = (2*c)/(-b - sqrt(b^2 - 4*a*c))
a*x^2 + b*x + c
```

## Machine epsilon

Machine epsilon is the distance between 1 and the next largest machine representable number.  It can be shown that the following simple code produces machine epsilon.

```matlab
a = 4/3;
b = a - 1;
c = b + b + b;
1-c
```

```matlab
s = 1.;
while 1+s > 1
    s = s/2
end
2*s
```
